### DATA SCIENCE

```python
# Configurar opciones del navegador
options = ChromeOptions()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
```

```python
driver = Chrome(options=options)
```

```python
driver.get('https://pe.indeed.com/jobs?q=data%20science&l=Lima&from=searchOnHP')
```

```python
# Listas para almacenar la información
companies = []
jobs = []
locations = []
dates = []
descriptions = []
observations_1 = []
observations_2 = []

while True:
    # Esperar hasta que aparezca al menos un elemento 'list'
    WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.CLASS_NAME, "cardOutline")))
    
    # Encontrar todos los elementos 'list'
    lists = driver.find_elements(By.CLASS_NAME, "cardOutline")

    # Recorrer cada 'list' y extraer la información requerida
    for list_element in lists:
        company = list_element.find_element(By.CLASS_NAME, "companyName").text
        job = list_element.find_element(By.CLASS_NAME, "jcs-JobTitle.css-jspxzf.eu4oa1w0").text
        location = list_element.find_element(By.CLASS_NAME, "companyLocation").text
        date = list_element.find_element(By.CLASS_NAME, "date").text
        description = list_element.find_element(By.CLASS_NAME, "job-snippet").text
        observation_1_elements = list_element.find_elements(By.CLASS_NAME, "urgentlyHiring")
        observation_1 = ""
        if observation_1_elements:
            for element in observation_1_elements:
                observation_1 += element.text
        observation_2_elements = list_element.find_elements(By.CLASS_NAME, "shelfItem.hiringMultipleCandidates")
        observation_2 = ""
        if observation_2_elements:
            for element in observation_2_elements:
                observation_2 += element.text
    
        # Agregar la información a las listas correspondientes
        companies.append(company)
        jobs.append(job)
        locations.append(location)
        dates.append(date)
        descriptions.append(description)
        observations_1.append(observation_1)
        observations_2.append(observation_2)
    
    # Verificar si existe el botón de siguiente página
    try:
        next_button = driver.find_element(By.XPATH, '//a[@data-testid="pagination-page-next"]')
        if 'disabled' in next_button.get_attribute("class"):
            break
    except NoSuchElementException:
        break
        
        
    time.sleep(random.uniform(3, 6))
    
    # Hacer clic en el botón de siguiente página
    driver.execute_script("arguments[0].click();", next_button)
    WebDriverWait(driver, 30).until(EC.staleness_of(lists[0]))

# Cerrar el navegador
driver.quit()
```

```python
# Crear un dataframe con la información recolectada
data_ds_in = {
    'Company': companies,
    'Job': jobs,
    'Location': locations,
    'Date': dates,
    'Description': descriptions,
    'Observation 1': observations_1,
    'Observation 2': observations_2
}

df_dsi = pd.DataFrame(data_ds_in)
df_dsi.index = df_dsi.index + 1
```

### Unir información de ambos portales

```python
info_v0=pd.concat([df_total_iv3,df_total_bv2],axis=0,keys=['INDEED', 'BUMERAN'])
```

```python
# Filtramos las publicaciones duplicadas entre ambos portales
# Se eliminarán los que tengan mayor antiguedad

def convert_to_days(text):
    if "días" in text:
        match = re.search(r'\d+', text)
        if match:
            return int(match.group())
    elif "ayer" in text:
        return 1
    elif "hoy" in text:
        return 0
    elif "Hoy" in text:
        return 0
    else:
        return -1

info_v0['Days'] = info_v0['Date'].apply(convert_to_days)

# Eliminar los registros duplicados basados en los campos 'ompany, Job y Days
info_v1 = info_v0.sort_values('Days').drop_duplicates(['Company', 'Job'], keep='last')
```

```python
info_v1.drop('Days', axis=1, inplace=True)
```

## Filtro de trabajos

```python
# Se excluyen los nombres de empleos que no es de nuestro interés
# A mayor detalle en el filtro, la información será más precisa con tu búsqueda

excluir = ["sr", "senior", "jefe", "gerente", ".net", "java", "sap", "lead", "practicante", "asesor", "desarrollador","ingeniero civil","químico"]
listax = "|".join(excluir)

# Filtrar los registros que contienen las palabras clave (ignorando mayúsculas y minúsculas)
info_v2 = info_v1[~info_v1["Job"].str.contains(listax, flags=re.IGNORECASE)]
```

```python
# Filtrar por años de experiencia requerida

excluir = ["4 años", "5 años", "6 años", "7 años", "8 años", "9 años", "10 años", "15 años", "20 años"]
listax = "|".join(excluir)
info_v3 = info_v2[~info_v2["Description"].str.contains(listax, flags=re.IGNORECASE,na=False)]
```

```python
# Filtrar por fecha de publicación

info_v4 = info_v3[~info_v3["Date"].str.contains("[6-9]|1[0-9]|2[0-9]|30")]
```

```python
info_v4 = info_v4.copy()
info_v4.loc[:, 'Order'] = 0

info_v4.loc[info_v4['Date'].str.contains('Hoy|Recién publicado'), 'Order'] = -1
info_v4.loc[info_v4['Date'].str.isnumeric(), 'Order'] = info_v4.loc[info_v4['Date'].str.isnumeric(), 'Date'].astype(int)
info_v4 = info_v4.sort_values(['Order', 'Date'], ascending=[True, True])
info_v4.drop('Order', axis=1, inplace=True)
```
