# EXTRACCIÓN DE INFORMACIÓN

## DATA SCIENCE

```python
# Configurar opciones del navegador
options = ChromeOptions()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
```

```python
driver = Chrome(options=options)
```

```python
driver.get('https://www.bumeran.com.pe/empleos-busqueda-ciencia-de-datos.html')
```

```python
companies = []
jobs = []
locations = []
dates = []
descriptions = []
observations_1 = []
observations_2 = []
options.add_argument("--headless")  # Ejecución en modo headless (sin interfaz gráfica)
service = Service("path/to/chromedriver")  # Reemplaza "path/to/chromedriver" con la ruta al archivo chromedriver.exe

# Inicializar el controlador de Selenium
driver = webdriver.Chrome(service=service, options=options)

while True:
    # Esperar hasta que aparezca al menos un elemento 'list'
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.TAG_NAME, "body")))
    
    # Encontrar todos los elementos 'list'
    lists = driver.find_elements(By.CLASS_NAME, "sc-csJRnm.cbWJbr")

    # Recorrer cada 'list' y extraer la información requerida
    for list_element in lists:
        company = list_element.find_element(By.CLASS_NAME, "sc-jkPxnQ.bxzQRW").text
        job = list_element.find_element(By.CLASS_NAME, "sc-eZXMBi.byiMiZ").text
        location = list_element.find_element(By.CLASS_NAME, "sc-dCVVYJ.hWcikA").text
        date = list_element.find_element(By.CLASS_NAME, "sc-kiXyGy.ebStiw").text
        description_elements = list_element.find_elements(By.CLASS_NAME, "sc-Ehqfj.ezOOTA")
        description = ""
        if description_elements:
            for element in description_elements:
                description += element.text
        
        # Agregar la información a las listas correspondientes
        companies.append(company)
        jobs.append(job)
        locations.append(location)
        dates.append(date)
        descriptions.append(description)

    # Buscar el enlace de "Avanzar" en la página actual
    next_button = driver.find_element(By.CSS_SELECTOR, "a.sc-hrBRpH.eGzIAQ")
    if 'disabled' in next_button.get_attribute("class"):
        break
    
    next_url = next_button.get_attribute("href")
    
    current_page = driver.current_url
    driver.get(next_url)
    new_page = driver.current_url
    if current_page == new_page:
        break

    time.sleep(random.uniform(3, 5))

driver.quit()
```

```python
# Crear un dataframe con la información recolectada
data_ds = {
    'Company': companies,
    'Job': jobs,
    'Location': locations,
    'Date': dates,
    'Description': descriptions
}

df_ds = pd.DataFrame(data_ds)

df_ds["Observation 1"] = np.nan
df_ds["Observation 2"] = np.nan
```

```python
campos_revisión = ['Company', 'Job', 'Location', 'Date']
df_ds1 = df_ds.drop_duplicates(subset=campos_revisión).reset_index(drop=True)
df_ds1.index = df_ds1.index + 1
```

## Filtro de trabajos

```python
df_total_v3 = df_total_v2[df_total_v2["Job"].str.contains("sr|Sr|SR|senior|Senior|SENIOR|jefe|Jefe|JEFE|gerente|Gerente|GERENTE|.net|.Net|.NET|java|Java|JAVA|SAP |lead|Lead|LEAD|practicante|Practicante|PRACTICANTE|asesor|Asesor|ASESOR|desarrollador|Desarrollador|DESARROLLADOR")==False]
```

```python
df_total_v3
```

```python
df_total_v4 = df_total_v3[df_total_v3["Description"].str.contains("4 años| 5 años| 6 años| 7 años| 8 años| 9 años| 10 años| 15 años")==False]
```

```python
df_total_v5 = df_total_v4[df_total_v4["Location"].str.contains("Lima|Perú")==True]
```

```python
df_total_v5
```

```python
df_total_v6 = df_total_v5[~df_total_v5["Date"].str.contains("[6-9]|1[0-9]|2[0-9]|30")]
```

```python
df_total_v6
```

```python
df_total_v6 = df_total_v6.copy()
df_total_v6.loc[:, 'Order'] = 0

df_total_v6.loc[df_total_v6['Date'].str.contains('Hoy|Recién publicado'), 'Order'] = -1
df_total_v6.loc[df_total_v6['Date'].str.isnumeric(), 'Order'] = df_total_v6.loc[df_total_v6['Date'].str.isnumeric(), 'Date'].astype(int)
df_total_v6 = df_total_v6.sort_values(['Order', 'Date'], ascending=[True, True])
df_total_v6.drop('Order', axis=1, inplace=True)
```
